```python
import pandas as pd
import folium
import json
```

```python
df = pd.read_csv('ccgtotalopiate.csv')
```

```python
m = folium.Map(location=[54.2, -2.45], zoom_start=5)
```

```python
m.choropleth(
    geo_path='./ccgsimple.json',
    data=df,
    columns=['CCG13CD', 'MG'],
    key_on='feature.properties.CCG13CD',
    fill_color='PuBu', 
    fill_opacity=0.7, 
    line_opacity=0.3,
    legend_name='Total opiates prescribed per person by CCG (mg) August 2010-February 2014'
)
```

```python
folium.LayerControl().add_to(m)
```
